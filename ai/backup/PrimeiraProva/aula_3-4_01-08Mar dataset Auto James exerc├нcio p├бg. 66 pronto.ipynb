O desvio padro  uma medida de disperso que indica o quanto os valores de um conjunto de dados esto dispersos em relao  mdia desse conjunto. 
Em outras palavras, o desvio padro quantifica a variabilidade ou a disperso dos valores em torno da mdia.

Clculo do desvio padro:
1) calcule a diferena entre cada valor do conjunto de dados e a mdia desse conjunto: xi - m
2) eleve essas diferenas ao quadrado (xi - m)^2
3) some esses valores quadrados
4) divida esta soma pela quantidade total de observaes
5) tire a raiz quadrada do resultado

```python
from sympy import symbols, sqrt, Sum, pprint
xi, mu = symbols('x_i mu')
N = symbols('N')
desvio_padrao = sqrt(Sum((xi - mu)**2, (xi, 1, N))/N)
pprint(desvio_padrao) # SymPy Pretty Print
```
